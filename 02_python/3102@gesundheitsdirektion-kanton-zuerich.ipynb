## Description

Der Datensatz enthält betriebliche Daten zu den Alters- und Pflegeheimen im Kanton Zürich. Er umfasst aggregierte Informationen zum Leistungsangebot der Heime, zu den erbrachten Leistungen, zum Personal sowie zu den Klienten der Einrichtungen. Die Daten werden jeweils im Rahmen der jährlichen Erhebung der Statistik der sozialmedizinischen Institutionen (SOMED) im Kanton Zürich erhoben. Der Inhalt der Erhebung ist seit Jahren stabil. Dies ermöglicht verlässliche Vergleiche über die Jahre hinweg. 
Die Daten beziehen sich ausschließlich auf die Alters- und Pflegeheime im Kanton Zürich und ermöglichen somit eine detaillierte Analyse der dortigen Versorgungssituation.
