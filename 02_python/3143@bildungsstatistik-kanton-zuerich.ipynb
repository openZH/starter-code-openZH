## Description

Ausgewiesen wird die Anzahl der Lernenden an Bildungsinstitutionen im Kanton Zürich und mit Wohnort im Kanton Zürich differenziert nach verschiedenen räumlichen Einheiten, sowie nach Schulstufe und Schultyp. Die Lernenden bis zur Schulstufe Sek II, sowie die Lernenden in der höheren Berufsbildung stammen aus der Statistik der Lernenden des Kantons Zürich, die Daten zu Fachhochschulen, pädagogische Hochschulen und Hochschulen stammen aus der Statistik zu den Studierenden und Abschlüssen an Hochschulen (SHIS-studex) des Bundesamtes für Statistik.
Die Daten werden ausgehend von den Daten auf Gemeindeebene für die darüberliegenden räumlichen Dimensionen aggregiert. Für Gemeinden die fusioniert haben werden die historischen Daten auf für die neuen Gemeinden aggregiert (Rückwärtsfusion).
