## Dataset
# **Arbeitslose und Stellensuchende in Zürcher Gemeinden und Bezirken**

## Description

Der Datensatz enthält monatliche Angaben zur Anzahl der Stellensuchenden und Arbeitslosen sowie die entsprechenden Quoten für alle Gemeinden und Bezirke im Kanton Zürich. 

In diesem Datensatz finden Sie detaillierte Informationen über die Arbeitsmarktsituation auf Gemeindeebene. Es wird zwischen Stellensuchenden und Arbeitslosen unterschieden: Als stellensuchend gelten jene Personen, die beim Regionalen Arbeitsvermittlungszentrum (RAV) gemeldet sind, unabhängig von ihrer sofortigen Verfügbarkeit (beispielsweise Personen in Kündigungsfrist oder Weiterbildung). Als arbeitslos werden jene Personen eingestuft, die keine Stelle haben, beim RAV gemeldet sind und sofort vermittelbar sind. Ergänzend wird die jeweilige Quote ausgewiesen, welche den prozentualen Anteil der Stellensuchenden und Arbeitslosen an der Erwerbsbevölkerung der jeweiligen Gemeinde darstellt. 

Die Daten stammen aus dem Informationssystem für die Arbeitsvermittlung und die Arbeitsmarktstatistik (AVAM). Dies ist das offizielle Datenerfassungssystem, in dem alle Personen registriert werden, die sich bei einem RAV anmelden.

Die Daten gelten als zuverlässig für Rückschlüsse auf die Arbeitslosigkeit in den Gemeinden. Bei der Nutzung ist jedoch zu beachten, dass eine Aggregation der Zahlen über mehrere Monate hinweg nicht zulässig ist, da es sich um Monatsbestände handelt und dieselben Stellensuchenden und Arbeitslosen mehrfach erfasst sein könnten. In Einzelfällen werden Personen, bei denen keine Standortangaben zur Gemeinde vorliegen, aus der Statistik entfernt; dabei handelt es sich monatsweit um eine kleine einstellige Anzahl an Personen. 

Die Daten beziehen sich auf das Gebiet des Kantons Zürich. Die Werte sind auf Ebene der Gemeinden ausgewiesen. Während die absoluten Zahlen der Stellensuchenden und Arbeitslosen pro Berichtsmonat auf Bezirksebene aggregiert werden können, ist dies für die Quoten aufgrund der notwendigen Gewichtung nicht ohne Weiteres möglich. Deshalb werden die aggregierten Quoten für die Bezirke als separater Datensatz ausgewiesen.
