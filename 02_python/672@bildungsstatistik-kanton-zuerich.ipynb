## Dataset
# **Gymnasiale Maturitätsquoten im Kanton Zürich**

## Description

Die gymnasiale Maturitätsquote beschreibt den häufigsten Zugangsweg zum Studium an einer Universität, Fachhochschule oder Pädagogischen Hochschule. Hier werden die durchschnittlichen gymnasialen Maturitätsquoten aller politischen Gemeinden, Bezirke und Raumplanungsregionen im Kanton Zürich ausgewiesen. Die Quoten werden pro räumlicher Einheit und Jahr wie folgt berechnet: Summe der Maturitätsabschlüsse von Personen mit Wohnsitz in der räumlichen Einheit in den vier letzten Jahren geteilt durch die Summe der Anzahl 19-jähriger Einwohnerinnen und Einwohner derselben räumlichen Einheit in den letzten vier Jahren. Die hier ausgewiesene Maturitätsquote wird aus einer Kombination der Statistik der Bildungsabschlüsse und der Bevölkerungsstatistik ermittelt.

Publiziert sind die gymnasiale Maturitätsquoten ab dem Gemeindestand von 2014 (Abschlüsse und Bevölkerungsstände von 2011 bis 2014).
